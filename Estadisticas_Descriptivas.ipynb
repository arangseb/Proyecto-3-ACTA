```python
materias = ['PUNT_INGLES', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES', 'PUNT_SOCIALES_CIUDADANAS', 'PUNT_LECTURA_CRITICA']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
axes = axes.flatten()  # Aplanar para facilitar el acceso a los subgráficos

for i, materia in enumerate(materias):
    ax = axes[i]
    ax.hist(
        df[materia],
        bins=10,
        color='lightsteelblue',
        edgecolor='black',
        alpha=0.8
    )
    ax.set_title(materia.replace('PUNT_', '').replace('_', ' ').title(), fontsize=14)
    ax.set_xlabel('Puntaje', fontsize=12)
    ax.set_ylabel('Frecuencia', fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.6)

if len(materias) < len(axes):
    for j in range(len(materias), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle('Distribución de Puntajes por Materias', fontsize=16, fontweight='bold', y=1.02)  # Título general
plt.show()

```
